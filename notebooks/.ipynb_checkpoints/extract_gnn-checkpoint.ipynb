```python
import sys
import mitsuba as mi
import drjit as dr
import numpy as np

mi.set_variant('scalar_rgb')

scene = mi.load_file('scenes/cornell-box/scene.xml')
film = scene.sensors()[0].film()
# sample_count = sampler.sample_count()
sample_count = 10
height, width = film.crop_size()
print(f"Film has size: [{height}, {width}]")
```

```python
""" Utils functions """
def mis_weight(pdf_a, pdf_b):
    
    pdf_a *= pdf_a;
    pdf_b *= pdf_b;
    w = pdf_a / (pdf_a + pdf_b);
    
    return dr.detach(dr.select(dr.isfinite(w), w, 0.))

def normalize(v):
    
    return v / np.sqrt(np.sum(v**2))

def has_flag(flag1, flag2):
    
    return (int(flag1) & int(flag2)) != 0

""" Prepare ray from sampled position """
def prepare_ray(pos, sensor, film, sampler, sample_count):
    
    diff_scale_factor = dr.rsqrt(sample_count)

    scale = mi.ScalarVector2f(1. / mi.ScalarVector2f(film.crop_size()))
    offset = -mi.ScalarVector2f(film.crop_offset()) * scale

    sample_pos = mi.Vector2f(pos + sampler.next_2d())
    adjusted_pos = (sample_pos * scale - offset)

    aperture_sample = mi.Point2f(.5)
    if sensor.needs_aperture_sample():
        aperture_sample = sampler.next_2d()

    time = sensor.shutter_open()
    if sensor.shutter_open_time() > 0.:
        time += sampler.next_1d() * sensor.shutter_open_time()

    wavelength_sample = 0. # by default

    (ray, ray_weight) = sensor.sample_ray_differential(time, 
                                                    wavelength_sample,
                                                    adjusted_pos,
                                                    aperture_sample)

    if (ray.has_differentials):
        ray.scale_differential(diff_scale_factor)
        
    return adjusted_pos, ray, ray_weight

""" Depth sampler from ray """
def sample(scene, sampler, ray_, medium, max_depth):
    
    # m_rr_depth: depth to begin russian roulette
    m_rr_depth = 5 # by default
    m_hide_emitters = False # Disable direct visibility of emitters if needed
    
    # ---------------- Context Loop variable -------------------
    ray = mi.Ray3f(ray_)
    throughput = mi.Spectrum(1.)
    result = mi.Spectrum(0.)
    eta = mi.Float(1)
    depth = mi.UInt32(0)
    
    valid_ray = not m_hide_emitters and dr.neq(scene.environment(), None)
    
    prev_si = dr.zeros(mi.Interaction3f)
    prev_bsdf_pdf = mi.Float(1.)
    prev_bsdf_delta = mi.Bool(True)
    bsdf_ctx = mi.BSDFContext()
    active = True
    
    while active:
        
        si = scene.ray_intersect(ray, mi.RayFlags.All, dr.eq(depth, 0))
        
        # ---------------------- Direct emission ----------------------
        if (dr.neq(si.emitter(scene), None)):
            
            ds = mi.DirectionSample3f(scene, si, prev_si)
            em_pdf = 0.;

            if (not prev_bsdf_delta):
                em_pdf = scene.pdf_emitter_direction(prev_si, ds, not prev_bsdf_delta)

                # Compute MIS weight for emitter sample from previous bounce
                mis_bsdf = mis_weight(prev_bsdf_pdf, em_pdf)

                # Accumulate result with throughput (bsdf)
                result += throughput * (ds.emitter.eval(si, prev_bsdf_pdf > 0.) * mis_bsdf)
                
        
        # Continue tracing the path at this point?
        active_next = (depth + 1 < max_depth) and si.is_valid()

        if not active_next:
            break # early exit for scalar mode

        bsdf = si.bsdf(ray)
        
        # ---------------------- Emitter sampling ----------------------

        # Perform emitter sampling?
        active_em = active_next and has_flag(bsdf.flags(), mi.BSDFFlags.Smooth)

        ds = dr.zeros(mi.DirectionSample3f)
        em_weight = dr.zeros(mi.Spectrum)
        wo = dr.zeros(mi.Vector3f)

        if active_em:
            
            # Sample the emitter
            (ds, em_weight) = scene.sample_emitter_direction(si, sampler.next_2d(), True, active_em)
            active_em &= dr.neq(ds.pdf, 0.)

            # Given the detached emitter sample, recompute its contribution
            # with AD to enable light source optimization.
            if dr.grad_enabled(si.p):
                
                ds.d = normalize(ds.p - si.p)
                em_val = scene.eval_emitter_direction(si, ds, active_em)
                em_weight = dr.select(dr.neq(ds.pdf, 0), em_val / ds.pdf, 0)

            wo = si.to_local(ds.d)
        
        # ------ Evaluate BSDF * cos(theta) and sample direction -------
        sample_1 = sampler.next_1d()
        sample_2 = sampler.next_2d()

        (bsdf_val, bsdf_pdf, bsdf_sample, bsdf_weight) = bsdf.eval_pdf_sample(bsdf_ctx, si, wo, sample_1, sample_2)

        # --------------- Emitter sampling contribution ----------------
        if active_em:
            bsdf_val = si.to_world_mueller(bsdf_val, -wo, si.wi)

            # Compute the MIS weight
            mis_em = dr.select(ds.delta, 1., mis_weight(ds.pdf, bsdf_pdf))

            # Accumulate, being careful with polarization (see spec_fma)
            result += throughput * (bsdf_val * em_weight * mis_em)
        
        # ---------------------- BSDF sampling ----------------------
        bsdf_weight = si.to_world_mueller(bsdf_weight, -bsdf_sample.wo, si.wi)

        ray = si.spawn_ray(si.to_world(bsdf_sample.wo))

        """When the path tracer is differentiated, we must be careful that
           the generated Monte Carlo samples are detached (i.e. don't track
           derivatives) to avoid bias resulting from the combination of moving
           samples and discontinuous visibility. We need to re-evaluate the
           BSDF differentiably with the detached sample in that case."""
        if dr.grad_enabled(ray):
            ray = dr.detach(ray)

            # Recompute 'wo' to propagate derivatives to cosine term
            wo_2 = si.to_local(ray.d)
            (bsdf_val_2, bsdf_pdf_2) = bsdf.eval_pdf(bsdf_ctx, si, wo_2, active)
            bsdf_weight[bsdf_pdf_2 > 0.] = bsdf_val_2 / dr.detach(bsdf_pdf_2)

        # ------ Update loop variables based on current interaction ------
        throughput *= bsdf_weight
        eta *= bsdf_sample.eta
        valid_ray |= active and si.is_valid() and \
                     not has_flag(bsdf_sample.sampled_type, mi.BSDFFlags.Null)

        # Information about the current vertex needed by the next iteration
        prev_si = si
        prev_bsdf_pdf = bsdf_sample.pdf
        prev_bsdf_delta = has_flag(bsdf_sample.sampled_type, mi.BSDFFlags.Delta)

        # -------------------- Stopping criterion ---------------------
        #dr.masked(depth, si.is_valid()) += 1
        depth += 1

        throughput_max = dr.max(throughput)
        
        rr_prob = dr.minimum(throughput_max * dr.sqr(eta), .95) + sys.float_info.epsilon
        rr_active = depth >= m_rr_depth
        rr_continue = sampler.next_1d() < rr_prob

        """Differentiable variants of the renderer require the the russian
           roulette sampling weight to be detached to avoid bias. This is a
           no-op in non-differentiable variants."""
        throughput[rr_active] *= dr.rcp(dr.detach(rr_prob))

        active = active_next and (not rr_active or rr_continue) and dr.neq(throughput_max, 0.)
        
    return result

def render_samples(pos, scene, n_samples, max_depth):
    
    sensor = scene.sensors()[0]
    film = sensor.film()
    sampler = sensor.sampler()
    medium = sensor.medium()
    
    final_result = dr.zeros(mi.Color3f)
    for _ in range(n_samples):
    
        adjusted_pos, ray, ray_weight = prepare_ray(pos, sensor, film, sampler, n_samples)
        result = sample(scene, sampler, ray, medium, max_depth)
        result *= ray_weight
        final_result += mi.Color3f(result)

    return final_result / mi.Float(n_samples)
```

```python
%timeit render_samples([0, 0], scene, 1000, 3)
```

```python
import copy
from functools import partial
from multiprocessing.pool import ThreadPool as Pool

# adapt native code to Mitsuba Python API for sampling each pos
image = np.zeros([width, height, 3])

n_samples = 100
# sample per pixel
for yy in range(height):

    y_row = np.full(
      shape=height,
      fill_value=yy
    )
    x_row = np.arange(width)

    pixels = list(zip(x_row, y_row))

    pool_obj = Pool()   
    pixel_row = pool_obj.map(
        partial(render_samples, scene=scene, n_samples=n_samples, max_depth=3),
        pixels
    )

    image[yy] = pixel_row

    print(f"Row n°{yy + 1} of {height}", end='\r')
```
